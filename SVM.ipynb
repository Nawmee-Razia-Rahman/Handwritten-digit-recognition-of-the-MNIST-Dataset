```python
import pandas as pd
import matplotlib.pyplot as plt, matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline
train=pd.read_csv("C:\\Users\\ASUS\\project3-2\\mnist_train.csv")
```

```python
test=pd.read_csv("C:\\Users\\ASUS\\project3-2\\mnist_test.csv")
```

```python
test
```

```python
images = train.iloc[0:60000,1:] 
labels = train.iloc[0:60000,:1]
```

```python
test_images = test.iloc[0:5000,1:]
test_labels = test.iloc[0:5000,:1]
```

```python
# we're splitting the training set here into four, the data for training and the data for validation

(trainData, valData, trainLabels, valLabels) = train_test_split(images, labels,
test_size=0.1, random_state=84)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_validate
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
import timeit
import time
```

```python
#algorithm accuracy/test accuracy
clf_svm = LinearSVC()
start=time.clock()
clf_svm.fit(trainData, trainLabels)
end=time.clock()
svm_cpu_time=end-start
y_pred_svm = clf_svm.predict(valData)
acc_svm = accuracy_score(valLabels, y_pred_svm) 
print("Validation accuracy:",acc_svm)
print("CPU calculating time:",svm_cpu_time)   

#now we are doing the same on the 'test data' 
test_data_prediction = clf_svm.predict(test_images)
test_data_acc = accuracy_score(test_labels, test_data_prediction)
print("Test data accuracy:",test_data_acc)

```

```python
val_error=(1-acc_svm)
print("Validation error:",val_error)
```

```python
test_error=(1-test_data_acc)
print("Test data Error:",test_error)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
cm = metrics.confusion_matrix(test_labels, test_data_prediction)
print(cm)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(test_data_acc)
plt.title(all_sample_title, size = 15);
```
